#### Important SOC categories after analyzing data 

```python
soc_categories = {
                    'EXECUTIVES' : ['CHIEF', 'EXECUTIVE', 'CEO', 'CFO'],
                  'ACTOR' : ['ACTOR', 'MUSIC'],
                  'MANAGERS' : ['MANAGER', 'MANAGER'],
                 'SCIENTIST' : ['SCIENTIST'],
                  'DATA' : ['DATA'],
                 'COMPUTER' : ['PROGRAMMER, COMPUTER', 'DEVELOPER', 
                              'INFORMATION', 'SOFTWARE'],
                  'EDUCATION' : ['EDUCATION', 'TEACHER', 'PROFESSOR',
                                 'ENGLISH', 'LANGUAGE', 'SCHOOL'],
                  'ENGINEER' : ['ENGINEER', 'TECHNICIAN', 'ELECTRONIC',
                                'ELECTRICAL', 'INDUSTRIAL', 'MECHANICAL'],
                  'PILOT' : ['PILOT'],
                  'MEDICAL' : ['DENTIST', 'MEDICAL', 'MENTAL', 'ORTHO',
                               'PHYSICIAN', 'RADIOLO'],
                  'MEDICINE-OTHERS' : ['PHARMACY', 'HEALTH', 'DIET'],
                  'DESIGNER' : ['DESIGNER'],
                  'DETECTIVE' : ['DETECTIVE'],
                  'BIOLOGIST' : ['BIO'],
                  'BUSINESS' : ['BUSINESS'],
                  'COOK' : ['COOK'],
                  'STATISTICIAN' : ['ANALYS', 'STATISTIC'],
                  'FINANCE' : ['FINANC', 'LOAN', 'SECURITY', 'ACCOUNTANT'],
                  'GEOLOGIST' : ['GEO'],
                  'SALES & MARKETING' : ['SALE', 'MARKET'],
                 } 
```

```python
#Category function to apply to SOC_NAME column
def find_category(unknown):
    unknown = unknown.upper()
    for key in soc_categories:
        for i in soc_categories[key]: 
            if(unknown.startswith(i) or unknown.find(i) != -1):
                return key
    return 'OTHER'
```

```python
df['SOC_NAME'] = df['SOC_NAME'].apply(find_category) #Apply function to categorize SOC_NAME category
```

```python
df_worksite = df.groupby('WORKSITE', as_index=False)['PREVAILING_WAGE'].median().sort_values(by='PREVAILING_WAGE')
#Get worksite dataframe to categorize states in 5 categories - 
# High income, Low income, Upper High income, Above average income and Average income states as we will see ahead
```

```python
mydict = dict(zip(df_worksite['WORKSITE'], df_worksite['PREVAILING_WAGE']))
```

```python
def normalise_row_wage(row):
    state = str(row['WORKSITE']).strip()
    state_median = mydict[state]
    if state_median <= row['PREVAILING_WAGE']:
        return 1
    return 0

# Add to column to dataframe to know if salary is above median as per state
df['WAGE_ABOVE_MEDIAN'] = df.apply(lambda row : normalise_row_wage(row), axis=1)
```

```python
def input_row_wage(state, wage):
    state_median = mydict[state]
    if state_median <= wage:
        return 1
    return 0
```

#### Made dictionary of types of states to categorize

```python
# After analysis, we found 5 states would serve the purpose better
high_income_states = dict(zip(df_worksite['WORKSITE'].iloc[-6:], df_worksite['PREVAILING_WAGE'].iloc[-6:]))
```

```python
high_income_states
```

```python
low_income_states = dict(zip(df_worksite['WORKSITE'].iloc[:9], df_worksite['PREVAILING_WAGE'].iloc[:9]))
```

```python
low_income_states
```

```python
average_income_states = dict(zip(df_worksite['WORKSITE'].iloc[9:25], df_worksite['PREVAILING_WAGE'].iloc[9:25]))
```

```python
average_income_states
```

```python
above_average_income_states = dict(zip(df_worksite['WORKSITE'].iloc[25:38], df_worksite['PREVAILING_WAGE'].iloc[25:38]))
```

```python
above_average_income_states
```

```python
upper_high_income_states = dict(zip(df_worksite['WORKSITE'].iloc[38:46], df_worksite['PREVAILING_WAGE'].iloc[38:46]))
```

```python
upper_high_income_states
```

```python
# Classify states through the dictionaries we created
def worksite_level_classifier(state):
    if state in high_income_states:
        return 'HIGH'
    elif state in low_income_states:
        return 'LOW'
    elif state in average_income_states:
        return 'AVERAGE'
    elif state in above_average_income_states:
        return 'ABOVE_AVERAGE'
    elif state in upper_high_income_states:
        return 'UPPER_HIGH'
    else:
        return 'UNDEFINED'
```

```python
df['STATE_TYPE'] = df['WORKSITE'].apply(worksite_level_classifier)
df = pd.get_dummies(df, columns=['WORKSITE'])
# Divide prevailing wage by median to avoid prevailing wage taking more weight in prediction and suppressing other components,
#as prevailing wage are much higher than other features. So, dividing it by median
wage_median = df['PREVAILING_WAGE'].median()
df['PREVAILING_WAGE'] = df['PREVAILING_WAGE'].apply(lambda x : x/wage_median)
```

#### Take only 100,000 approved and all denied cases, this is done because of high imbalance in data (Only 3% denial from 3 million records), so this will balance the data

```python
df_zeros = df.loc[df['CASE_STATUS'] == 0]
```

```python
df_ones = df.loc[df['CASE_STATUS'] == 1]
```

```python
df_ones_to_process = df_ones.sample(n=100000, random_state=25)
```

```python
frames = [df_zeros, df_ones_to_process]
```

```python
result = pd.concat(frames) #Finally we get the data ready to give it to machine learning
```

#### Implement Logistic Regression using Sklearn, just as a measure for comparison

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
X = result.ix[:,1:].values
y = result.ix[:,0].values
# normalized_X = preprocessing.normalize(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
LogReg = LogisticRegression()
LogReg.fit(X_train, y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
